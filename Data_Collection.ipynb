```python
if 'mapping' not in globals():
    mapping = {}

import math
import matplotlib.pyplot as plt
import seaborn

# สมมติว่ามี data DataFrame และ mapping อยู่แล้ว
obj_cols = data.select_dtypes(include=['object']).columns
n = len(obj_cols)
rows = math.ceil(n/3)
fig, axes = plt.subplots(rows, 3, figsize=(12, 4*rows))
axes = axes.flatten()

for i, col in enumerate(obj_cols):
    s = data[col]
    # แปะ label หากมี mapping
    if col in mapping:
        labels_series = s.map(mapping[col]).fillna(s).astype(str)
    else:
        labels_series = s.astype(str)

    counts = labels_series.value_counts(dropna=False)

    # สีสำหรับ 2–3 หมวด (Yes/No/สามระดับ) ตามที่คุณชอบ
    if counts.size <= 3:
        colors = seaborn.color_palette('pastel', counts.size)
    else:
        colors = seaborn.color_palette('Paired', counts.size)

    # --- ส่วนที่แก้ไข ---
    # 1. เก็บผลลัพธ์ (patches) ไว้
    # 2. ลบ 'labels=counts.index' ออก
    patches, texts, autotexts = axes[i].pie(counts.values,
                                            autopct='%1.1f%%',
                                            startangle=90,
                                            counterclock=False,
                                            colors=colors)
    
    axes[i].set_title(col)

    # 3. เพิ่ม Legend (กล่องคำอธิบายสี)
    axes[i].legend(patches, 
                   counts.index, 
                   title="Categories",  
                   loc="best")          


for j in range(i+1, len(axes)):
    axes[j].axis('off')

plt.tight_layout()
plt.show()
```

```python
# เลือกเฉพาะคอลัมน์ที่มีจริงใน df

n = len(data.select_dtypes(include=['number']).columns)

# จัดกริดให้พอดีอัตโนมัติ
ncols = 3
nrows = math.ceil(n / ncols)

fig, axes = plt.subplots(nrows, ncols, figsize=(5*ncols, 4*nrows), sharex=False, sharey=False)
axes = axes.flatten()

# เตรียมโทนสี (จะเอาจาก palette มาก็ได้ แต่ส่งเข้าเป็น color ทีละกราฟ)
palette = seaborn.color_palette('GnBu', max(n, 3))

for i, var in enumerate(data.select_dtypes(include=['number']).columns):
    seaborn.histplot(
        x=data[var].dropna(),
        ax=axes[i],
        kde=True,
        bins=30,
        color=palette[i % len(palette)],
        edgecolor='white',
        linewidth=0.5
    )
    axes[i].set_title(var)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# ปิดแกนที่เกินมา (ถ้าช่องในกริดมากกว่าจำนวนคอลัมน์)
for j in range(i + 1, len(axes)):
    axes[j].axis('off')

fig.suptitle(
    "Histogram Plot of Numerical Variables",
    fontsize=20, fontweight='bold', fontfamily='serif'
)
fig.subplots_adjust(top=0.92)  # กันหัวไม่ชนกราฟ
plt.show()
```
