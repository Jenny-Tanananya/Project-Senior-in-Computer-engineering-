```python
import pandas
import numpy
import seaborn
import matplotlib.pyplot as plt
import sklearn.preprocessing
import sklearn.model_selection
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pandas.read_csv('../data/heart_disease.csv')
data
```

```python
data = data.dropna()
data
```

```python
numeric_features = data.select_dtypes(include=['number']).columns
def cap_outliers(data, lower_percentile=0.01, upper_percentile=0.99):
    df_copy = data.copy()
    for col in numeric_features:
        lower_bound = df_copy[col].quantile(lower_percentile)
        upper_bound = df_copy[col].quantile(upper_percentile)
        df_copy[col] = df_copy[col].clip(lower=lower_bound, upper=upper_bound)
    return df_copy

numeric_features = cap_outliers(data[numeric_features])
numeric_features
```

```python
plt.figure(figsize=(12, 0.6*len(numeric_features.columns)+2))
seaborn.boxplot(data=numeric_features, orient='h', showfliers=True)
plt.title("Boxplots (outliers)")
plt.tight_layout()
plt.show()
```

```python
categorical_features = data.select_dtypes(include=['object', 'category']).columns
categorical_features
```

```python
categorical_features.isnull().sum()
```

```python
onehot_features = [
    'Gender', 
    'Smoking', 
    'Family Heart Disease', 
    'Diabetes', 
    'High Blood Pressure', 
    'Low HDL Cholesterol', 
    'High LDL Cholesterol', 
    
    ]

onehot_mapping = [
    ('Male', 'Female'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
   
]

ordinal_features = [
    'Exercise Habits',
    'Alcohol Consumption',
    'Stress Level',
    'Sugar Consumption'
]

ordinal_mapping = [
     ('Low', 'Medium', 'High'),
    ('Low', 'Medium', 'High'),
    ('Low', 'Medium', 'High'),
    ('Low', 'Medium', 'High')
]

mapping = {
    'Heart Disease Status': {
        'No': 0,
        'Yes': 1
    }
}
```

```python
onehot_encoder = sklearn.preprocessing.OneHotEncoder(categories=onehot_mapping, sparse_output=False)
onehot_encoder
```

```python
onehot_encoder.fit(data[onehot_features])
```

```python
onehot_encoder.get_feature_names_out()
```

```python
data_onehot_features = pandas.DataFrame(
        data = onehot_encoder.transform(data[onehot_features]),
        columns = onehot_encoder.get_feature_names_out(),
        index = data.index  
    )
data_onehot_features
```

```python
ordinal_encoder = sklearn.preprocessing.OrdinalEncoder(categories=ordinal_mapping)
```

```python
ordinal_encoder.fit(data[ordinal_features])
```

```python
ordinal_encoder.get_feature_names_out()
```

```python
data_ordinal_features = pandas.DataFrame(
        data = ordinal_encoder.transform(data[ordinal_features]),
        columns = ordinal_encoder.get_feature_names_out(),
        index = data.index  
    )
data_ordinal_features
```

```python
mapping = pandas.DataFrame(
    data = data.replace(mapping),
    columns = ['Heart Disease Status'],
    index = data.index
)
mapping
```

```python
categorical_features = (
    data_ordinal_features
    .join(data_onehot_features)
    
)
categorical_features
```

```python
data =(
    numeric_features
    .join(categorical_features)
)
data
```

```python
scaler = RobustScaler()
numeric_features_scaled = pandas.DataFrame(
    scaler.fit_transform(data),
    columns=data.columns,
    index=data.index
)

print("\nStatistics after scaling:")
print(numeric_features_scaled.describe())

# Box plot เปรียบเทียบก่อนและหลัง scaling
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# ก่อน scaling
seaborn.boxplot(data=data, orient='h', ax=ax1)
ax1.set_title("Before Scaling")

# หลัง scaling
seaborn.boxplot(data=numeric_features_scaled, orient='h', ax=ax2)
ax2.set_title("After Scaling")

plt.tight_layout()
plt.show()
```

```python
# สร้าง scalers
robust_scaler = RobustScaler()
minmax_scaler = MinMaxScaler()

# Scale ข้อมูลด้วยทั้งสองวิธี
robust_scaled = pandas.DataFrame(
    robust_scaler.fit_transform(data),
    columns=data.columns,
    index=data.index
)

minmax_scaled = pandas.DataFrame(
    minmax_scaler.fit_transform(data),
    columns=data.columns,
    index=data.index
)

# เปรียบเทียบผลลัพธ์ด้วย box plots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 12))

# ข้อมูลดั้งเดิม
seaborn.boxplot(data=data, orient='h', ax=ax1)
ax1.set_title("Original Data")

# RobustScaler
seaborn.boxplot(data=robust_scaled, orient='h', ax=ax2)
ax2.set_title("After RobustScaler")

# MinMaxScaler
seaborn.boxplot(data=minmax_scaled, orient='h', ax=ax3)
ax3.set_title("After MinMaxScaler")

plt.tight_layout()
plt.show()

# แสดงสถิติพื้นฐานของทั้งสองวิธี
print("\nStatistics after RobustScaler:")
print(robust_scaled.describe())
print("\nStatistics after MinMaxScaler:")
print(minmax_scaled.describe())
```

```python
data_final = data.join(mapping)
```

```python
x = data_final.drop('Heart Disease Status', axis=1)
y = data_final['Heart Disease Status']
```

```python
data_train_x, data_test_x, data_train_y, data_test_y = sklearn.model_selection.train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    shuffle=True
)
```

```python
data_train = pandas.concat([data_train_x, data_train_y], axis=1)
print('Data Train:')
print(data_train)
```

```python
data_test = pandas.concat([data_test_x, data_test_y], axis=1)
print('Data Test:')
print(data_test)
```

```python
data_test.to_csv('../data/heart_disease_drop_preprocessed_test.csv', index=False)
```
