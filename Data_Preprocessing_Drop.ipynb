## Data Preprocessing Drop

```python
import pandas
import numpy
import seaborn
import matplotlib.pyplot as plt
import sklearn.preprocessing
import sklearn.model_selection
```

```python
data = pandas.read_csv('/Users/tanananyathongkum/Documents/Visual Studio Code/Project_CE/data/heart_disease.csv')
data
```

```python
data = data.dropna()
data
```

```python
numeric_features = data.select_dtypes(include=['number']).columns
def cap_outliers(data, lower_percentile=0.01, upper_percentile=0.99):
    df_copy = data.copy()
    for col in numeric_features:
        lower_bound = df_copy[col].quantile(lower_percentile)
        upper_bound = df_copy[col].quantile(upper_percentile)
        df_copy[col] = df_copy[col].clip(lower=lower_bound, upper=upper_bound)
    return df_copy

numeric_features = cap_outliers(data[numeric_features])
numeric_features
```

```python
plt.figure(figsize=(12, 0.6*len(numeric_features.columns)+2))
seaborn.boxplot(data=numeric_features, orient='h', showfliers=True)
plt.title("Boxplots (outliers)")
plt.tight_layout()
plt.show()
```

```python
categorical_features = data.select_dtypes(include=['object', 'category']).columns
categorical_features
```

```python
categorical_features.isnull().sum()
```

```python
onehot_features = [
    'Gender', 
    'Smoking', 
    'Family Heart Disease', 
    'Diabetes', 
    'High Blood Pressure', 
    'Low HDL Cholesterol', 
    'High LDL Cholesterol', 
    
    ]

onehot_mapping = [
    ('Male', 'Female'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
    ('No', 'Yes'),
   
]

ordinal_features = [
    'Exercise Habits',
    'Alcohol Consumption',
    'Stress Level',
    'Sugar Consumption'
]

ordinal_mapping = [
     ('Low', 'Medium', 'High'),
    ('Low', 'Medium', 'High'),
    ('Low', 'Medium', 'High'),
    ('Low', 'Medium', 'High')
]

mapping = {
    'Heart Disease Status': {
        'No': 0,
        'Yes': 1
    }
}
```

```python
onehot_encoder = sklearn.preprocessing.OneHotEncoder(categories=onehot_mapping, sparse_output=False)
onehot_encoder
```

```python
onehot_encoder.fit(data[onehot_features])
```

```python
onehot_encoder.get_feature_names_out()
```

```python
data_onehot_features = pandas.DataFrame(
        data = onehot_encoder.transform(data[onehot_features]),
        columns = onehot_encoder.get_feature_names_out(),
        index = data.index  
    )
data_onehot_features
```

```python
ordinal_encoder = sklearn.preprocessing.OrdinalEncoder(categories=ordinal_mapping)
```

```python
ordinal_encoder.fit(data[ordinal_features])
```

```python
ordinal_encoder.get_feature_names_out()
```

```python
data_ordinal_features = pandas.DataFrame(
        data = ordinal_encoder.transform(data[ordinal_features]),
        columns = ordinal_encoder.get_feature_names_out(),
        index = data.index  
    )
data_ordinal_features
```

```python
mapping = pandas.DataFrame(
    data = data.replace(mapping),
    columns = ['Heart Disease Status'],
    index = data.index
)
mapping
```

```python
categorical_features = (
    data_ordinal_features
    .join(data_onehot_features)
    .join(mapping)
)
categorical_features
```

```python
data =(
    numeric_features
    .join(categorical_features)
)
data
```

```python
x = data.drop('Heart Disease Status', axis=1)
y = data['Heart Disease Status']
```

```python
data_train_x, data_test_x, data_train_y, data_test_y = sklearn.model_selection.train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=0,
    shuffle=True
)
```

```python
MinMaxScaler = sklearn.preprocessing.MinMaxScaler(feature_range=(-1, 1), copy=True)
MinMaxScaler
```

```python
MinMaxScaler.fit(data_train_x)
```

```python
data_train_x = pandas.DataFrame(
        data = MinMaxScaler.transform(data_train_x),
        columns = data_train_x.columns,
        index = data_train_x.index
    )
data_train_x
```

```python
data_test_x = pandas.DataFrame(
        data = MinMaxScaler.transform(data_test_x),
        columns = data_test_x.columns,
        index = data_test_x.index
    )
data_test_x
```

```python
data_train = pandas.concat([data_train_x, data_train_y], axis=1)
print('Data Train:')
print(data_train)
```

```python
data_test = pandas.concat([data_test_x, data_test_y], axis=1)
print('Data Test:')
print(data_test)
```
